```python
xo=list(range(0,10))

br=['#', '-', '-', '-']
```

```python
from IPython.display import clear_output
```

```python
def print_board():   
   
    print( ' | ' + xo[7] + ' | ' + xo[8] + ' | ' + xo[9] + ' | ' )
    print( ' | ' + br[1] + ' | ' + br[2] + ' | ' + br[3] + ' | ' )
    print( ' | ' + xo[4] + ' | ' + xo[5] + ' | ' + xo[6] + ' | ' )
    print( ' | ' + br[1] + ' | ' + br[2] + ' | ' + br[3] + ' | ' )
    print( ' | ' + xo[1] + ' | ' + xo[2] + ' | ' + xo[3] + ' | ' )
```

```python
def reset_board():
    xo[1]=xo[2]=xo[3]=xo[4]=xo[5]=xo[6]=xo[7]=xo[8]=xo[9]=' '
    print_board()
```

```python
def choose_xy():
    
    choose=True
    global player1
    global player2
    
    while choose:
        
        a=input('Player 1 - Choose X or O: ')
        
        if a.upper()=='X' or a.upper()=='O':
            choose=False
            
            if a.upper()=='X':
                global player1
                global player2
                player1='X'
                player2='O'
                print('\nPlayer 1 is X \nPlayer 2 is O \nLET´S PLAY!')
            else:
                player1='O'
                player2='X'
                print('\nPlayer 1 is O \nPlayer 2 is X \nLET´S PLAY!')
        
        else:
            continue
```

```python
def player1_input():
    
     
    p1=True
    
    while p1:
        print('\n')
        b=input('Player1 choose between 1 to 9 to select your playing spot:\n\n | 7 | 8 | 9 |\n | 4 | 5 | 6 |\n | 1 | 2 | 3 |\n\n ')
        
        if b.isdigit():
            c=int(b)
        else:
            continue
    
        if c in range(1,10) and xo[c]==' ':
            p1=False
                     
            xo[c]=player1
        
            print('\n')
            print_board()
        
        else:
            print('That spot is already chosen, please select another one')
            continue
```

```python
def player2_input():
    
    
    p2=True    
    
    while p2:
        print('\n')
        b=input('Player2 choose between 1 to 9 to select your playing spot:\n\n | 7 | 8 | 9 |\n | 4 | 5 | 6 |\n | 1 | 2 | 3 |\n\n ')
        
        if b.isdigit():
            c=int(b)
        else:
            continue
    
        if c in range(1,10) and xo[c]==' ':
            p2=False
                     
            xo[c]=player2
        
            print('\n')
            print_board()
        
        else:
            print('That spot is already chosen, please select another one')
            continue
```

```python
def check_win_player2():
    
        
    if xo[9]==xo[5]==xo[1]==player2 or xo[1]==xo[2]==xo[3]==player2 or xo[4]==xo[5]==xo[6]==player2 \
    or xo[7]==xo[8]==xo[9]==player2 or xo[7]==xo[5]==xo[3]==player2 or xo[7]==xo[4]==xo[1]==player2 \
    or xo[8]==xo[5]==xo[2]==player2 or xo[9]==xo[6]==xo[3]==player2:
            
            repeat=input('\nPlayer 2 has won! Would you like to play again? Type YES if you wanna play again ')
            if repeat.upper()=='YES':
                reset_board()
                clear_output()
                print_board()
                choose_xy()
            else:
                return 'Goodbye!'
                                               
    else:
        return 'verdadero'
```

```python
def check_tie():
    
     
    if xo[1]!=' ' and  xo[2]!=' ' and xo[3]!=' ' and xo[4]!=' ' and xo[5]!=' ' and xo[6]!=' ' and xo[7]!=' ' \
    and xo[8]!=' ' and xo[9]!=' ':
        
        repeat=input('\nIt´s a tie! Would you like to play again? Type YES if you wanna play again ')
        
        if repeat.upper()=='YES':
                reset_board()
                clear_output()
                print_board()
                choose_xy()
        else:
            return 'Goodbye!'
                                    
    else:
        return 'verdadero'
```
